## Create a function that says if a review is subjective or objective

```python
def getsubjectivity(contain):
    return TextBlob(contain).sentiment.subjectivity
```

## Create a function that provides the main sentiment of each review

```python
def getpolarity(contain):
    return TextBlob(contain).sentiment.polarity
```

## Create a column called Subjectivity taking as input the score of subjectivity of each review
## Note that values are between 0 and 1

```python
train_df["Subjectivity"]=train_df["review"].apply(getsubjectivity)
```

## Create a column called Subjectivity taking as input the score of polarity of each review
## Note that values are between -1 and 1

```python
train_df["Polarity"]=train_df["review"].apply(getpolarity)
```

```python
# Schow the new train data
train_df
```

## generate each word in the review

```python
allWords="".join([j for j in train_df["review"]])
```

## Use the wordcloud function to generate all words appearing the high occurence

```python
WORDSCLOUD=WordCloud(width=500, height=300,random_state=21, max_font_size=119,stopwords = set(STOPWORDS))
```

```python
# generate all mostly repeated words
WORDSCLOUD.generate(allWords)
```

```python
# print all the wordcloud
plt.imshow(WORDSCLOUD,interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Depending on the polarity score, creating a function assigning a negative, Neutral or positive sentiment to each score.

```python
def getAnalysis(score):
    if score<0:
        return "Negative"
    elif score==0:
        return "Neutral"
    else:
        return "Positive"
```

## Creating a column namely Analysis and taking as input the sentiment

```python
train_df["Analysis"]=train_df["Polarity"].apply(getAnalysis)
```
